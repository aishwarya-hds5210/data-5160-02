```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the csv file and skip the first 4 rows as it contains the header
data = pd.read_csv("CO2_emission.csv", skiprows = 4)
data
```

```python
# Check the structure of dataframe
data.shape
```

```python
# Column names
data.columns
```

```python
# calculate the average pollutants across the years from 1960 to 2022 by removing the individual year columns
year_columns = data.loc[:, '1960':'2022']
average_pollutants = year_columns.mean(axis=1)
data['pollutants'] = average_pollutants
data.drop(columns=year_columns.columns, inplace=True)
data
```

```python
# Filter the DataFrame where the 'pollutants' values are greater than 550,000
highest = data[data['pollutants'] > 550000]
highest
```

```python
# Create a bar plot to visualize CO2 emission across ccountries
plt.figure(figsize=(12, 6))
plt.bar(highest['Country Code'], highest['pollutants'])

plt.xlabel('Country Code')
plt.ylabel('Pollutants')
plt.title('CO2 emission')
plt.xticks(rotation='vertical')
plt.tight_layout()

plt.show()
```

```python
# Read the csv file and skip the first 4 rows as it contains the header
data2 = pd.read_csv("Life_expentency_world.csv", skiprows = 4)
data2
```

```python
# Check the structure of dataframe
data2.columns
```

```python
# Calculate the row-wise average for the selected columns and create a new column 'avg_life_exp'
year_columns = data2.loc[:, '1960':'2022']
data2['avg_life_exp'] = year_columns.mean(axis=1)
data2.drop(columns=year_columns.columns, inplace=True)
data2
```

```python
highest2 = data2[data2['avg_life_exp'] > 75]
highest2
```

```python
# Plotting scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(highest2['Country Code'], highest2['avg_life_exp'], color='red', alpha=0.7)
plt.title('Average Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Read the csv file and skip the first 4 rows as it contains the header
data3 = pd.read_csv('ATM.csv', skiprows = 4)
data3
```

```python
# Check the structure of dataframe
data3.columns
```

```python
# calculate the average of the 'ATMs' year-wise and create new column named 'ATMs'
year_columns = data3.loc[:, '1960':'2022']
data3['ATMs'] = year_columns.mean(axis=1)
data3.drop(columns=year_columns.columns, inplace=True)
data3
```

```python
# Filter rows where the 'ATMs' values are greater than 90.
highest3 = data3[data3['ATMs'] > 90]
highest3
```

```python
# Plot line chart showing distribution of ATMs
plt.figure(figsize=(10, 6))
plt.plot(highest3['Country Code'], highest3['ATMs'], marker='o', color='green', linestyle='-', linewidth=2, markersize=8, label='ATMs')
plt.title('Number of ATMs in various country')
plt.xlabel('Country')
plt.ylabel('ATMs')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
