```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
data = pd.read_csv("Racial_and_Social_Equity_Composite_Index_Current.csv")
data
```

```python
# Remove duplicate rows
data = data.drop_duplicates()
data
```

```python
# Dropping unnecessary columns
data = data.drop(columns = ["OBJECTID",	"GEOID", "PCT_POP_UNDER200PCT_POVERTY", "PTL_POP_UNDER200PCT_POVERTY", "POP_ADULT_CIVILIAN_NONINST", "PCT_ADULT_MENTALHLTHNOTGOOD", "PTL_ADULT_MENTALHLTHNOTGOOD", "PCT_LESS_BACHELOR_DEGREE", "PTL_LESS_BACHELOR_DEGREE", "POP_ENGLISH_LANG_LEARNERS", "POP_5_YRS_AND_OLDER", "POP_LESS_THAN_BACHELOR_DEG", "PCT_FOREIGN_BORN", "PTL_FOREIGN_BORN", "LIFE_EXPECTANCY_AT_BIRTH", "PTL_LOW_LIFE_EXPECT_AT_BIRTH", "POP_25_YRS_AND_OLDER", "POP_POV_STATUS_DETERMINED",	"POP_FOREIGN_BORN",	"POP_TOTAL_FOR_FOREIGN_BORN", "POP_UNDER200PCT_POVERTY",	"POP_ADULT_WITH_DISABILITY", "PCT_ENGLISH_LANG_LEARNERS", "PTL_ENGLISH_LANG_LEARNERS", "PCT_PEOPLE_OF_COLOR", "PTL_PEOPLE_OF_COLOR", "POP_ADULT_CIVILIAN_NONINST", "SHAPE_Area", "SHAPE_Length"])
data
```

```python
# Since there are 27 columns, it is better to choose required columns from the dataset and create as new table
new_data = data[["PCT_ADULT_NOLEISUREPHYSACTIV", "PCT_ADULT_WITH_OBESITY",
                 "SOCIOECON_DISADV_SCORE", "HEALTH_DISADV_SCORE",
                 "SOCIOECON_DISADV_QUINTILE", "PCT_ADULT_WITH_DISABILITY", "ACRES_TOTAL", "POP_TOTAL_FOR_POC"]]

new_data
```

```python
# Rename columns in the new_data DataFrame
new_data.rename(columns={
    'PCT_ADULT_NOLEISUREPHYSACTIV': 'No_phy_activity',
    'PCT_ADULT_WITH_OBESITY': 'Obesity',
    'PCT_ADULT_WITH_DISABILITY': 'Disability'
}, inplace=True)
new_data
```

```python
# Update the column which are facing highest disadvantage.
new_data.loc[new_data["SOCIOECON_DISADV_QUINTILE"] == 'Highest Equity Priority/Most Disadvantaged']
```

```python
# creating a calculated field
new_data['Disability'] = new_data['Disability'].apply(lambda x: x * 100)
new_data
```

```python
# Rounding off the decimal values to two digits
new_data['Disability'] = new_data['Disability'].round(2)
new_data
```

```python
# Dealing with NULL or Infinite Values
new_data.dropna(inplace=True)
new_data
```

```python
# Creating a Calculated Field
# Create a new column 'PopulationDensity' based on division of 'POP_TOTAL' by 'ACRES_TOTAL'
new_data['PopulationDensity'] = new_data['POP_TOTAL_FOR_POC'] / data['ACRES_TOTAL']
new_data
```

```python
# Plot boxplot for Population Density
plt.figure(figsize=(7, 4))
y_axis_range = [0, 100]
plt.ylim(y_axis_range)
new_data['PopulationDensity'].plot(kind='box')
plt.ylabel('Values')
plt.title('Boxplot')
plt.show()
```

```python
from scipy.stats.mstats import winsorize

# Apply Winsorizing to 'PopulationDensity' to handle outliers
new_data['PopulationDensity'] = winsorize(new_data['PopulationDensity'], limits=[0.01, 0.01])

plt.figure(figsize=(7, 4))
y_axis_range = [0, 100]
plt.ylim(y_axis_range)
new_data['PopulationDensity'].plot(kind='box')
plt.ylabel('Values')
plt.title('Boxplot (After Winsorizing)')
plt.show()
```

```python
import numpy as np

# Apply log transformation to 'PopulationDensity'
new_data['PopulationDensity'] = np.log1p(new_data['PopulationDensity'])

plt.figure(figsize=(7, 4))
y_axis_range = [0, 10]
plt.ylim(y_axis_range)
new_data['PopulationDensity'].plot(kind='box')
plt.ylabel('Values')
plt.title('Boxplot (After Log Transformation)')
plt.show()
```

```python
# Save the cleaned dataset to a CSV file
new_data.to_csv('cleaned_dataset.csv', index=False)
```
